```python
import pandas as pd 
import datetime as dt 
from twitterscraper import query_tweets
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from langdetect import detect 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.mlab as mlab
import re

def detector(x):
    try:
       return detect(x)
    except:
        None 
        
analyzer = SentimentIntensityAnalyzer()

#pd.set_option('display.max_colwidth', -1)
```

```python
begin_date = dt.date(2019,1,1)
end_date = dt.date(2019,5,15)

limit = 10000

tweets = query_tweets("#VoteForIndia OR #IndiaVote OR #IndiaVotes OR #IndianElection2019 OR #BJP OR #Congress \
OR #NarendraModi OR #RahulGandi OR #PMModi OR #LokSabhaElections2019 OR #AAP OR #AamAadmiParty", \
                      begindate=begin_date, enddate=end_date, limit=limit)
```

```python
df = pd.DataFrame(t.__dict__ for t in tweets)

df[['fullname','likes', 'replies', 'retweets', 'text', 'timestamp', 'user']]
```

```python
#filter for english tweets only
df['lang'] = df['text'].apply(lambda x:detector(x))
df = df[df['lang'].isin(['en','hi'])]

#removed duplicates because of sponsored tweets? 
df.drop_duplicates(subset = 'text',inplace = True)

#save in files for future analysis
df.to_csv('original_tweets_data.csv')
df.to_json('original_tweets_data.json')
```

```python
# This block of code is to clean our data and get ready for further analysis
# since twitterscraper API took care of any NaN values we don't need to worry about it.
'''couple of things that we can do while cleaning data
1. convert all english text to lower case
2. remove all puntuations
3. remove all stopwords
4. make sure you should not remove '#' charachter from the text
5. Need to understand how can we avoid tweets that are not related to Indian election. For example, #congress hashtag is used across many countries. 
6. this user "ModiRahulModi" has tweeted an irrelevant tweet on 2019-03-21. Looks like some fake tweet. Is there any way this can be avoided.
7. Need to implement tokenization. I think we can use textblob in our case.
8. Need to implement 
9. Understand and implement N-grams
10. Understand and implement WordCloud
11. Then perform sentiment analysis
12. Then plot various graphs to explain and visualize your analysis.
13. See how can we implement spaCy in this project. probably in future works
'''
# Convert all the string to lower cases
df['selftext'] = df['selftext'].str.lower()
# below code is to remove all hyperlinks
df['selftext'] = df['selftext'].apply(lambda x: re.sub('http\S*', '', x))
# below code is to remove multiple spaces with single space
df['selftext'] = df['selftext'].apply(lambda x: re.sub('\s+', ' ', x))
# below code is to remove all the punctuation
df['selftext'] = df['selftext'].apply(lambda x: re.sub('[^\w\s]', '', x))
# below code is to remove all english and hindi stop words
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['selftext'] = df['selftext'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
# below code will be used to perform Lemmatizer
from nltk import WordNetLemmatizer
```

```python
#get sentiment scores
sentiment_score = selectDF['text'].apply(lambda x: analyzer.polarity_scores(x))
#put sentiment score into dataframe
sentimentDF = pd.concat([selectDF, sentiment_score.apply(pd.Series)],1)
#sentimentDF
```
