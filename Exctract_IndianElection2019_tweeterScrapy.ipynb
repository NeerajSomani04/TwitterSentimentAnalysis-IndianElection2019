```python
import pandas as pd 
#import datetime as dt 
#from datetime 
import datetime #timedelta
from twitterscraper import query_tweets
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from langdetect import detect 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.mlab as mlab

def detector(x):
    try:
       return detect(x)
    except:
        None 
        
analyzer = SentimentIntensityAnalyzer()
```

```python
begin_date = datetime.date(2019,1,1)
final_df = pd.DataFrame([])

for i in range(142):
    new_begin = begin_date + datetime.timedelta(days=i)
    new_end = begin_date + datetime.timedelta(days=i+1)
    #print(new_begin, ' ', new_end)
    tweets = query_tweets("#VoteForIndia OR #IndiaVote OR #IndiaVotes OR #IndianElection2019 OR #BJP OR #Congress OR #NarendraModi OR #RahulGandi OR #PMModi OR #LokSabhaElections2019 OR #AAP OR #AamAadmiParty", begindate = new_begin, enddate= new_end, limit = 10000)
    temp_df = pd.DataFrame(t.__dict__ for t in tweets)
    final_df = final_df.append(temp_df, ignore_index = True)
```
